```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
plt.plot(x,y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Example')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#OO 
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes2.plot(y,x)
```

```python
#Does the same thing as figure_axes
fig,axes = plt.subplots(nrows = 1,ncols = 2)
for cur_ax in axes:
    cur_ax.plot(x,y)    
    
```

```python
fig,axes = plt.subplots(nrows = 1,ncols = 2)
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
fig = plt.figure(figsize = (3,2))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,label = 'xy')
ax.plot(y,x,label = 'yx') 
ax.legend(loc = 0)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color= 'blue', lw = 3)
ax.plot(y,x,color= 'green', linestyle= '--')
ax.plot(y,y,color= 'red',marker= '>',markersize = 10)
ax.set_xlim([0,15])
```
