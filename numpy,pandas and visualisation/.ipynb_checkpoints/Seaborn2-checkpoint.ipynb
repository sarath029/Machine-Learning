```python
import seaborn as sns 
%matplotlib inline
```

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
tc = flights.corr()
```

```python
sns.heatmap(tc,annot = True)
```

```python
fp = flights.pivot_table(index= 'month',columns='year',values='passengers')
sns.heatmap(fp)
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_lower(sns.kdeplot)
g.map_upper(sns.swarmplot)
```

```python
g = sns.FacetGrid(data = iris,col = 'species' ,row = 'petal_width')
g.map(sns.distplot,'petal_length')
```
