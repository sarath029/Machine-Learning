```python
# Seaborn Visualisation 
```

```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
 sns.distplot(tips['total_bill'],kde = False)
```

```python
 sns.distplot(tips['tip'])
```

```python
sns.jointplot('tip','total_bill',tips)
```

```python
sns.jointplot('tip','total_bill',tips,kind = 'hex')
```

```python
sns.jointplot('tip','total_bill',tips, kind = 'reg')
```

```python
sns.jointplot('tip','total_bill',tips,kind = 'kde')
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue = 'sex',palette = 'coolwarm')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
#categorical Plots
sns.barplot(x = 'sex',y = 'tip',data = tips)
```

```python

sns.boxplot(x = 'day',y = 'tip',data = tips,hue = 'smoker')
```

```python

sns.countplot('day',data = tips)
```

```python
sns.violinplot('day','tip',data = tips,hue = 'sex', split = True)
```

```python
sns.boxplot(x = 'day',y = 'tip',data = tips)
```

```python
sns.scatterplot(x = 'day',y = 'tip',data = tips)
```

```python
sns.swarmplot(x = 'day',y = 'tip',data = tips)
```

```python
sns.factorplot(x = 'day',y = 'tip',data = tips,kind = 'bar')
```
